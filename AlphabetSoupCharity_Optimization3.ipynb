## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow import keras


#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
columns = ["EIN", "NAME", "STATUS"]
application_df = application_df.drop(columns,1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE
binning_df = application_df["APPLICATION_TYPE"].value_counts()
binning_df
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
application_types_to_replace = binning_df[binning_df<1000].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
classes_df = application_df["CLASSIFICATION"].value_counts()
classes_df
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
classes_df[classes_df > 3]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classifications_to_replace = classes_df[classes_df < 100].index
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_df_num = pd.get_dummies(application_df)
application_df_num.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_df_num["IS_SUCCESSFUL"].values
X = application_df_num.drop(["IS_SUCCESSFUL"],1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# # Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
# #  YOUR CODE GOES HERE
number_input_features = len(X_train_scaled[0])
nn = tf.keras.models.Sequential()
```

```python
# Optimize method 1: update the model

nn.add(tf.keras.layers.Dense(units = 60,
                            activation = "relu",
                            input_dim = number_input_features
                            ))

# nn.add(tf.keras.layers.Dense(units = 80,
#                             activation = "relu"
#                             ))

nn.add(tf.keras.layers.Dense(units = 80,
                            activation = "relu"
                            ))

nn.add(tf.keras.layers.Dense(units = 20,
                            activation = "relu"
                            ))

# nn.add(tf.keras.layers.Dense(units = 15,
#                             activation = "relu"
#                             ))

nn.add(tf.keras.layers.Dense(units = 1,
                             activation="sigmoid"
                             ))

nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss = "binary_crossentropy",
           optimizer = "adam",
           metrics=["accuracy"]
           )
```

```python
# Train the model
#  YOUR CODE GOES HERE
history = nn.fit(X_train_scaled, y_train,
                 epochs=200 #callbacks=callbacks,

                 )

# fit(X_train_scaled, y_train, epochs=500)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
#  YOUR CODE GOES HERE
nn.save('AlphabetSoupCharity_Optimization3.h5')
```
